# 분포의 시각화

```python
# 샘플데이터 (1~100 사이의 랜덤 정수 1000개)
import numpy as np
data = np.random.randint(1,101,1000)
```

## 히스토그램

```python
plt.hist(data)
plt.show()
# 데이터의 분포를 볼 수 있다.
```

```python
plt.hist(data, bins = 100) # bins = 100 이면 100개로 쪼개서 보여준다.
plt.show()
```

## 상자수염 그래프
- boxplot

```python
plt.boxplot(data)
plt.show()
# 맨 밑의 막대는 데이터의 최솟값
# 맨 위의 막대는 데이터의 최댓값
# 중간의 막대는 중간값(50%)
# 막대의 아랫부분은 25%
# 막대의 윗부분은 75%
```

## 바이올린 그래프
- violinplot

```python
plt.violinplot(data)
plt.show()

# 두께로 데이터의 분포를 확인할 수 있다.
```

# 색상, 마커, 선

## 색상
- color = 색상
- 색상 이름을 사용한다. https://matplotlib.org/2.0.2/examples/color/named_colors.html
- 자주 사용되는 색깔은 약자를 사용할 수 있다.
- (blue:b, green:g, red:r, cyan:c, magenta:m, yellow:y, black:k, white:w)
- hex code를 사용할 수 있다. (16진수로 되어있는 색상 코드)

```python
x = ['봄','여름','가을','겨울']
y = [19,30,15,-10]
```

```python
# 색상이름
plt.plot(x,y, color = 'skyblue')
```

```python
# 색상이름 약자 (blue:b, green:g, red:r, cyan:c, magenta:m, yellow:y, black:k, white:w)
plt.plot(x,y, color = 'r')
```

```python
# 색상코드 -> 인터넷에서 '색상코드표' 
plt.plot(x,y, color = '#A566FF')
```

## 마커
- marker : 마커종류 (* . , o v ^ < > 1 2 3 4 s p * h H + x D d)
- markersize, ms : 마커사이즈
- markeredgecolor, mec : 마커 선 색깔
- markeredgewidth, mew : 마커 선 굵기
- markerfacecolor, mfc : 마커 내부 색깔

```python
plt.plot(x,y, color = '#A566FF', marker='*')
```

```python
plt.plot(x,y, color = '#A566FF', marker='o', ms='10', mec='b', mew='3', mfc='y')
# 마커 약자로 사용하는 것이 편하다
```

## 선
- linestyle == ls : 선 스타일
*    '-' solid line style
*    '--' dashed lin style
*    '-.' dash-dot line style
*    ':' dotted line style

- linewidth == lw : 선 굵기

```python
plt.plot(x,y, color = '#A566FF', marker='o', ms='10', mec='b', mew='3', mfc='y'
        , ls=':', lw = 3)
```

## [색상][마커][선]
- [color][marker][line]
- 이 순서로 ' ' 안에 넣어주면 된다. 각각의 표현만으로 색상만, 색상 선만, 마커 선만 가능

### 색상

```python
plt.plot(x,y,'r')
```

### 마커

```python
plt.plot(x,y,'o')
```

### 선

```python
plt.plot(x,y,'--')
```

### 색상, 선

```python
plt.plot(x,y,'y--')
```

### 색상 마커 선

```python
plt.plot(x,y,'ro--')
```

```python
plt.plot(x,y,'ro--', ms=10, mec='b', mfc='y', lw=3)
```
