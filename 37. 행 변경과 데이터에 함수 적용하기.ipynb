- 데이터프레임의 끝에 행추가

# 인덱스를 지정하여 추가/수정하기
- 데이터프레임.loc[인덱스] = 추가할 데이터
- 데이터프레임.loc[인덱스] = 수정할 데이터
- 인덱스가 존재하면 해당 인덱스의 데이터가 수정된다.
- 인덱스가 존재하지 않으면 데이터프레임의 끝에 데이터가 추가된다.

```python
# 인덱스 35에 추가
df.loc[35] = ['aaa',70,80,90]

# 인덱스 34에 추가
df.loc[34] = ['bbb',80,90,100]
```

```python
# 인덱스 30
df.loc[30] = ['ccc',60,70,80]
```

# 데이터 준비하기

```python
df = pd.read_csv('data/scores.csv')
df = df.head()
df_copy = df.copy()
```

```python
df.head()
```

```python
df_copy.head()
```

# 함수로 컬럼의 데이터 변경하기
- 컬럼.apply(함수명)
- 컬럼.apply(함수명,매개변수명=매개변수값)
-- 매개변수명을 명시해주어야한다.
- 적용할 함수가 미리 정의되어 있어야 한다.

```python
# df['math']의 모든 점수에 5점 더하기.(--> 105점?)
df_copy.math = df.math+5
df_copy
```

```python
# df['math']의 모든 점수에 5점 더하기. 100점이 넘을 수 없다.
def plus5(x):
    score = x+5
    if score >= 100:
        score = 100
    return score
df_copy['math'] = df['math'].apply(plus5)
df_copy
```

```python
# df['eng']의 모든 점수에 n점 더하기. 100점이 넘을 수 없다.
def plusn(x,n):
    score = x+n
    if score >= 100:
        score = 100
    return score
df_copy['eng'] = df['eng'].apply(plusn,n=1)
df_copy
```

# 함수로 행/열의 데이터 집계하기
- 데이터프레임.apply(함수명,axis=0):열단위로 함수가 적용됨
- 데이터프레임.apply(함수명,axis=1):행단위로 함수가 적용됨

```python
df = pd.read_csv('data/scores.csv')
df = df.head()
df.index = df.name
df.drop(columns=['name'], inplace=True)
df_copy = df.copy()
```

```python
df_copy.head()
```

```python
# 행단위 데이터 출력
def print_me(x0):
    print(x0)
df.apply(print_me,axis=1)
```
