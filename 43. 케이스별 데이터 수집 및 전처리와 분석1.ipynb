```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 그래프를 노트북 안에 그리기 위해 설정
%matplotlib inline

# 필요한 패키지와 라이브러리를 가져온다.
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False

# 폰트 지정하기
plt.rcParams['font.family'] = 'Gulim' # -> 내 컴퓨터의 '굴림'체를 가져옴
```

```python
df = pd.read_csv('data/서울시 코로나19 확진자 현황2.csv')
```

```python
df.drop(columns = ['환자번호','국적','환자정보','이동경로','등록일','수정일','노출여부','조치사항'], inplace=True)
```

```python
df['확진일'] = pd.to_datetime(df['확진일'])
```

```python
df['지역'] = df['지역'].str.strip() # strip()함수는 공백을 제거한다.
```

```python
df['지역'] = df['지역'].astype('category')
```

```python
df.isnull().sum()
```

```python
## 여기까지 42강에서 43강을 위해 요약적으로 가져옴..
df
```

# 구 별 확진자 동향

```python
# # 우선 데이터를 보고 어떻게 집계를 할 것인지 생각..
# 확진일 -> 인덱스
# 지역 -> 컬럼
# 연번 -> 추가 확진자수로 value값..
```

## 확진일-구별 확진자수 집계

### 피봇테이블 만들기

```python
df_gu = pd.pivot_table(df, index='확진일', columns='지역', values='연번', aggfunc='count', margins=True)
df_gu
```

### 서울시 일별 확진자 동향 분석

```python
s_date = df_gu['All'][:-1] # 맨밑의 'All'인덱스는 가져오지 않기위해..
s_date # 시리즈로 가져오기
```

```python
# 서울시 일별 추가확진자가 많았던 순으로 보기
s_date.sort_values(ascending = False) # 내림차순으로 보기. (디폴트는 오름차순)
# 21년9월24일에 확진자가 제일 많았다는 것을 볼 수 있다.
```

```python
# 서울시 일별 추가확진자 시각화

# x축 -> 날짜(확진일) y축 -> 추가확진자의 수
x = s_date.index
y = s_date.values
plt.plot(x,y) # 일자별 수를 볼때 꺾은 선 그래프를 많이 사용한다.
plt.title('서울시 일별 추가확진자(~2021.09.28)')
plt.xlabel('확진일')
plt.ylabel('추가확진자 수')
plt.xticks(rotation = 45) # x축의 이름을 45도로 회전 # 겹치기 때문..
plt.show()
```

### 서울시 구별 누적확진자 비교


```python
df_gu
```

```python
# 서울시 구별 누적확진자 많은 순으로 보기
s_gu = df_gu.loc['All'][:-1] # 마지막 전체 합계 All을 빼기.
s_gu = s_gu.sort_values(ascending = False)
s_gu
```

```python
# 서울시 구별 누적확진자 많은 순으로 시각화

# # 시각화를 어떻게 할지 생각.. 
# x -> 지역 이름
# y -> 누적확진자 수
# 막대그래프가 보기에 좋겠지..?

x = s_gu.index
y = s_gu.values

plt.figure(figsize=(10,7)) # 그래프의 사이즈 설정 가로10,세로7
plt.barh(x,y) # barh 눕힌 막대그래프
plt.title('서울시 구별 누적확진자(~2021.9.28)', size=20)
plt.show()
```
