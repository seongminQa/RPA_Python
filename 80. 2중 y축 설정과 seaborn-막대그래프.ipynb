2중 y축을 표시하여 2가지 유형의 데이터를 한번에 표현할 수 있다.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False

plt.rcParams['font.size'] = 12
```

# 샘플데이터
* 어떤 학생의 나이에 따른 키와 몸무게의 변화

```python
age = [13,14,15,16,17]
height = [160,165,170,173,177]
weight = [80,85,83,78,73]
```

# 두가지 정보를 하나의 그래프에 그리기
* 축을 분리하기 위해 객체지향으로 그린다.
* fig, ax = plt.subplots()

```python
fig, ax = plt.subplots() # ()안엔 행렬식으로 들어간다
ax.bar(age, height, color='skyblue', width=0.5, ec='lightgray', label='height')
ax.plot(age, weight, color='darkred', marker='o', ls='-.', label='weight')
plt.legend()

plt.show()
```

## 2중 y축 만들기
* x축을 공유하는 새로운 axes객체를 만든다.<br>
`axes객체.twinx()`

```python
fig, ax1 = plt.subplots() # ()안엔 행렬식으로 들어간다
ax1.bar(age, height, color='skyblue', width=0.5, ec='lightgray', label='height')

ax2 = ax1.twinx()
ax2.plot(age, weight, color='darkred', marker='o', ls='-.', label='weight')

plt.show()
```

## 축 레이블 표시하기
`axes객체.set_xlabel(x레이블)`<br>
`axes객체.set_ylabel(y레이블)`<br>

```python
fig, ax1 = plt.subplots() # ()안엔 행렬식으로 들어간다
ax1.bar(age, height, color='skyblue', width=0.5, ec='lightgray', label='height')

ax2 = ax1.twinx()
ax2.plot(age, weight, color='darkred', marker='o', ls='-.', label='weight')

ax1.set_xlabel('나이')
ax1.set_ylabel('키(cm)')
ax2.set_ylabel('몸무게(kg)')

plt.show()
```

## y축 범위 지정
* axes객체.set_ylim(y축눈금범위)

```python
fig, ax1 = plt.subplots() # ()안엔 행렬식으로 들어간다
ax1.bar(age, height, color='skyblue', width=0.5, ec='lightgray', label='height')

ax2 = ax1.twinx()
ax2.plot(age, weight, color='darkred', marker='o', ls='-.', label='weight')

ax1.set_xlabel('나이')
ax1.set_ylabel('키(cm)')
ax2.set_ylabel('몸무게(kg)')

ax1.set_ylim(150, 180)
ax2.set_ylim(60, 90)

plt.show()
```

## y축 눈금
* axes객체.set_yticks(y축눈금)
* axes객체.tick_params(...)

```python
fig, ax1 = plt.subplots() # ()안엔 행렬식으로 들어간다
ax1.bar(age, height, color='skyblue', width=0.5, ec='lightgray', label='height')

ax2 = ax1.twinx()
ax2.plot(age, weight, color='darkred', marker='o', ls='-.', label='weight')

ax1.set_xlabel('나이')
ax1.set_ylabel('키(cm)')
ax2.set_ylabel('몸무게(kg)')

ax1.set_ylim(150, 180)
ax2.set_ylim(60, 90)

ax1.set_yticks(height) # y값 가지고 있는 것만 표현
ax2.set_yticks(weight)

ax1.tick_params(axis='y', colors='skyblue')
ax2.tick_params(axis='y', colors='darkred')

plt.show()
```

## 범례 표시
* axes객체별로 legend메소드 호출

```python
fig, ax1 = plt.subplots() # ()안엔 행렬식으로 들어간다
ax1.bar(age, height, color='skyblue', width=0.5, ec='lightgray', label='height')

ax2 = ax1.twinx()
ax2.plot(age, weight, color='darkred', marker='o', ls='-.', label='weight')

ax1.set_xlabel('나이')
ax1.set_ylabel('키(cm)')
ax2.set_ylabel('몸무게(kg)')

ax1.set_ylim(150, 180)
ax2.set_ylim(60, 90)

ax1.set_yticks(height) # y값 가지고 있는 것만 표현
ax2.set_yticks(weight)

ax1.tick_params(axis='y', colors='skyblue')
ax2.tick_params(axis='y', colors='darkred')

ax1.legend() # 객체 지향적 그래프 그리기..
ax2.legend()

plt.show()
```

## 그리드 표시
* axes객체.grid()

```python
fig, ax1 = plt.subplots() # ()안엔 행렬식으로 들어간다
ax1.bar(age, height, color='skyblue', width=0.5, ec='lightgray', label='height')

ax2 = ax1.twinx()
ax2.plot(age, weight, color='darkred', marker='o', ls='-.', label='weight')

ax1.set_xlabel('나이')
ax1.set_ylabel('키(cm)')
ax2.set_ylabel('몸무게(kg)')

ax1.set_ylim(150, 180)
ax2.set_ylim(60, 90)

ax1.set_yticks(height) # y값 가지고 있는 것만 표현
ax2.set_yticks(weight)

ax1.tick_params(axis='y', colors='skyblue')
ax2.tick_params(axis='y', colors='darkred')

ax1.legend()
ax2.legend()

ax1.grid(axis='y', ls='--', color='skyblue')
ax2.grid(axis='y', ls='--', color='red')

plt.show()
```

seaborn으로 막대그래프를 그릴 수 있다.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False
```

# 샘플데이터

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

# 요일별 팁 평균

## matplotlib으로 시각화

### 데이터 가공
* 요일별 팁 평균 계산
* 그룹핑 : 데이터프레임.groupby(그룹기준컬럼)[통계적용컬럼].통계함수

```python
day_tip_mean = tips.groupby(['day'])['tip'].mean()

# tips.groupby(['day']).agg(mean=('tip','mean'))
```

### 데이터 시각화
* plt.bar(x,y)

```python
plt.bar(day_tip_mean.index, day_tip_mean)

plt.xlabel('day')
plt.ylabel('tip')
plt.title('요일별 팁 평균', size=15)

plt.show()
```

## seaborn으로 그리기
* <b>`sns.barplot(data=데이터프레임명, x=x축컬럼, y=y축컬럼)`</b>
* x축데이터로 그룹핑한 y축데이터의 평균값을 계산하여 그래프를 그려준다.
* 신뢰구간(CI:Confidence Interval)을 함께 표시<br>

```python
sns.barplot(data=tips, x='day', y='tip', ci=None)
plt.title('요일별 팁 평균', size=15)
plt.show()

# 알아서 평균값을 내어 보여줌
# 자동으로 막대 별 색깔이 다름
```

# 요일별 팁 합계

## matplotlib으로 그리기

```python
day_tip_sum = tips.groupby('day')['tip'].sum()
day_tip_sum
```

```python
plt.bar(day_tip_sum.index, day_tip_sum)
plt.xlabel('day')
plt.ylabel('tip')
plt.title('요일별 팁 합계', size=15)
plt.show()
```

## seaborn으로 그리기
* estimator = 통계함수

```python
sns.barplot(data=tips, x='day', y='tip', ci=None, estimator=sum)
# 별도의 데이터 가공을 하지 않고 바로 그릴 수 있다.

plt.title('요일별 팁 합계', size=15)
plt.show()
```

# 요일별 팁 합계를 흡연여부로 비교
* hue : y를 그룹핑할 컬럼

```python
sns.barplot(data=tips, x='day', y='tip', ci=None, estimator=sum
           , hue='smoker')

plt.title('요일별 팁 합계', size=15)
plt.show()
```

* hue 색상 변경 : palette = 구분:색상 딕셔너리

```python
sns.barplot(data=tips, x='day', y='tip', ci=None, estimator=sum
           , hue='smoker', palette = {'Yes':'grey', 'No':'skyblue'})

plt.title('요일별 팁 합계', size=15)
plt.show()
```

* pyplot의 메소드로 그리드 추가

```python
sns.barplot(data=tips, x='day', y='tip', ci=None, estimator=sum
           , hue='smoker', palette = {'Yes':'grey', 'No':'skyblue'})

plt.title('요일별 팁 합계', size=15)
plt.grid(axis='y', ls=':')

plt.show()
```
