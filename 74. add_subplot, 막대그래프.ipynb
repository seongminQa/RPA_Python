전체 행 열과 그래프 순서를 지정하여 서브플롯을 그릴 수 있다.

```python
import matplotlib.pyplot as plt
```

# 데이터 불러오기

```python
import seaborn as sns
anscombe = sns.load_dataset('anscombe')
df1 = anscombe[anscombe['dataset']=='I']
df2 = anscombe[anscombe['dataset']=='II']
df3 = anscombe[anscombe['dataset']=='III']
df4 = anscombe[anscombe['dataset']=='IV']
```

```python
df4
```

# 전체 행 열과 그래프 순서에 따라 서브플롯 그리기
1) figure 객체를 생성한다.<br>
    <b>`fig=plt.figure()`</b><br><br>
2) 서브플롯을 그릴 axes 객체를 생성한다.<br>
    <b>`ax = fig.add_subplot(전체행개수,전체열개수,순서)`</b><br><br>


</img>

3) axes 객체에 그래프를 그린다.<br>
<b>`ax.plot(x,y)`</b>

4) 축 공유하기 :  어떤 axes의 축을 공유할 것인지 지정한다.<br> 
<b>`sharex=axes객체, sharey=axes객체`<br>

```python
# 1) figure 객체를 생성한다.
fig = plt.figure(figsize=(9,6), facecolor='ivory')

# 2) 서브플롯을 그릴 axes 객체를 생성한다.
# 4) 축 공유하기 : 어떤 axes의 축을 공유할 것인지 지정한다.
ax1 = fig.add_subplot(2,2,1) # 2행, 2열 중에 첫번째!
ax2 = fig.add_subplot(2,2,2, sharex=ax1, sharey=ax1)
ax3 = fig.add_subplot(2,2,3, sharex=ax1, sharey=ax1)
ax4 = fig.add_subplot(2,2,4)
# 전체 x축 y축을 ax1의 범위로 통일 (share)

# 3) axes 객체에 그래프를 그린다.
# 5) 범례 표시하기
ax1.plot(df1['x'], df1['y'], 'o', label='ax1')
ax2.plot(df2['x'], df2['y'], '^', label='ax2')
ax3.plot(df3['x'], df3['y'], '*', label='ax3')
ax4.plot(df4['x'], df4['y'], 'd', label='ax4')
ax1.legend(loc=2)
ax2.legend(loc=2)
ax3.legend(loc=2)
ax4.legend(loc=2)

# 4) 틱 변경하기
ax4.set_xticks(range(1,20,1))
ax4.set_yticks(range(1,15,1))

# 6) figure 제목 추가하기
fig.suptitle('Anscombe', fontsize=20) # 그냥 size 해도 된다.

# 8) 그래프 크기,간격 최적화하기
fig.tight_layout()

plt.show()
```

[학습목표]<br>
막대그래프를 그리고, 막대그래프의 여러가지 옵션을 지정할 수 있다.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 그래프에 한글 설정
plt.rc('font',family='Malgun Gothic')

# 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False
```

# 샘플데이터 
* 어느 두 매장의 요일별 평균 매출액

```python
df1 = pd.DataFrame({'요일':['월','화','수','목','금','토','일'], 
                   '매출액':[10000,9000,11000,8000,13000,15000,14000]})
df1
```

```python
df2 = pd.DataFrame({'요일':['월','화','수','목','금','토','일'], 
                   '매출액':[9000,9500,13000,7000,12000,14000,11000]})
df2
```

# 막대그래프
* plt.bar(x축데이터,y축데이터)
* plt.barh(x축데이터,y축데이터)

## 세로 막대 그래프

```python
plt.bar(df1['요일'],df1['매출액'])
plt.show()
```

## 가로 막대 그래프

```python
plt.barh(df2['요일'],df2['매출액'])
plt.show()
```

# 막대 폭 지정

## 세로 막대 그래프
* width = 0~1사이의 실수(default:0.8)

```python
plt.bar(df1['요일'],df1['매출액'], width=0.4)
plt.show()
```

## 가로막대 그래프
* height = 0~1사이의 실수(default:0.8)

```python
plt.barh(df2['요일'],df2['매출액'], height=0.4)
plt.show()
```

# 막대 색상 지정

```python
plt.bar(df1['요일'],df1['매출액'], width=0.4, color='r')
plt.show()
```

## 막대마다 다른 색 지정
* ['r','orange','y','g','b','navy','violet']

```python
plt.bar(df1['요일'],df1['매출액'], width=0.4,
        color=['r','orange','y','g','b','navy','violet'])
# 막대의 갯수만큼 컬러리스트를 만들어준다.

plt.show()
```

# 막대 테두리
* edgecolor = 테두리 색상
* linewidth = 테두리 두께

```python
plt.bar(df1['요일'],df1['매출액'], width=0.4,
        color=['r','orange','y','g','b','navy','violet'],
       edgecolor='gray', linewidth=2)
# 막대의 갯수만큼 컬러리스트를 만들어준다.

plt.show()
```

# 막대 패턴 지정하기

## 막대에 패턴 지정
* hatch 파라미터에 기호 전달 : '/', '\', '|', '-', '+', 'x', 'o', 'O', '.', '*'

```python
plt.bar(df1['요일'],df1['매출액'], width=0.4,
        color=['r','orange','y','g','b','navy','violet'],
       edgecolor='gray', linewidth=2, hatch='x')
# 막대의 갯수만큼 컬러리스트를 만들어준다.

plt.show()
```

## 패턴의 밀도 지정
* 패턴기호의 개수로 밀도를 조정한다.

```python
plt.bar(df1['요일'],df1['매출액'], width=0.4,
        color=['r','orange','y','g','b','navy','violet'],
       edgecolor='gray', linewidth=2, hatch='xxx')
# 막대의 갯수만큼 컬러리스트를 만들어준다.

plt.show()
```

## 막대마다 다른 패턴 지정
* 막대그래프를 객체로 받는다.
* 막대마다 다른 패턴을 지정한다.
* set_hatch(기호)

```python
bars = plt.bar(df1['요일'],df1['매출액'], width=0.4,
        color=['r','orange','y','g','b','navy','violet'],
       edgecolor='gray', linewidth=2, hatch='xxx')
bars[0].set_hatch('*')
bars[1].set_hatch('+')
bars[2].set_hatch('/')
bars[3].set_hatch('-')
bars[4].set_hatch('|')
bars[5].set_hatch('*')
# 막대의 갯수만큼 컬러리스트를 만들어준다.

plt.show()
```

# 막대 위치 지정

<b>`align = center/edge`</b><br>
* 디폴트: center<br>
* edge로 지정하면 막대의 왼쪽 끝과 틱을 맞춘다.<br>
* 막대의 오른쪽 끝과 틱을 맞추려면 width를 음수로 지정한다.<br>    

```python
plt.bar(df1['요일'],df1['매출액'], width=0.4, align='edge')
plt.show()
```

```python
plt.bar(df1['요일'],df1['매출액'], width=-0.4, align='edge')
plt.show()
```

## 두 개의 막대그래프 비교하기

```python
plt.bar(df1['요일'],df1['매출액'])
plt.bar(df2['요일'],df2['매출액'])
```

```python
# df1과 df2의 막대그래프 비교
plt.bar(df1['요일'],df1['매출액'], width=0.4, align='edge')
plt.bar(df2['요일'],df2['매출액'], width=-0.4, align='edge')

plt.title('두 매장의 요일별 매출액 평균', size=20)
plt.grid(axis='y', ls=':')
```

```python
# df1과 df2의 막대그래프 비교
plt.barh(df1['요일'],df1['매출액'], height=0.4, align='edge')
plt.barh(df2['요일'],df2['매출액'], height=-0.4, align='edge')

plt.title('두 매장의 요일별 매출액 평균', size=20)
plt.grid(axis='x', ls=':')
```
