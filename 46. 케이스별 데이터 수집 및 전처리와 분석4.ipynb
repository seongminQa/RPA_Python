# 삼겹살 가격 분석

## 삼겹살 데이터
- 특정 문자열이 포함되어있는지 확인 : .str.contains(문자열)

```python
# 2021-06 데이터 이용 -> 날짜와 가격의 상관관계

df['품목 이름']=='삼겹살'
df[df['품목 이름'].str.contains('삼겹살')]
df_sam = df[ (df['품목 이름'].str.contains('삼겹살')) & (df['년도-월']=='2021-06') &(df['실판매규격'].str.contains('600g')) ]
df_sam
# 컬럼의 값이 '문자형(object)' 일 때, str.함수 를 사용할 수 있다.**
```

```python
# 삼겹살 600g의 평균 가격은?
# df_sam.groupby('시장/마트 이름').agg(가격=('가격(원)','mean'))
df_sam['가격(원)'].mean()
```

```python
# 삼겹살 600g의 최고 가격은?
df_sam['가격(원)'].max()
```

```python
# 삼겹살 600g의 최저 가격은?
df_sam['가격(원)'].min()
```

```python
# 최저가격이 너무 낮다 잘못된 데이터인지 확인
df_sam[df_sam['가격(원)']<5000]
```

## 우리동네 삼겹살 가격

```python
gu = input('구이름: ')
df_sam_gu = df_sam[df_sam['자치구 이름'] == gu][['시장/마트 이름','품목 이름','실판매규격','가격(원)']].drop_duplicates()
df_sam_gu
```

```python
# 시각화
x = df_sam_gu['시장/마트 이름']
y = df_sam_gu['가격(원)']
plt.scatter(x,y)
plt.title(gu + ' 삼겹살 가격')
plt.grid(True) # x,y축에 줄을 그어 보기에 더 편하게 해준다.
plt.show()
```

## 승차/하차 테이블 분리

### 승차 테이블 만들기

```python
df.head(1)
# 필요한 컬럼 정하기
# 사용월, 호선명, 지하철역
```

```python
# 공통 컬럼
df_in1 = df.iloc[:,0:3]

# 승차 컬럼만 가져오기
df_in2 = df.iloc[:,3::2]
df_in2.columns = df_in2.columns.str.split(' ').str[0]
df_in2.columns # 시간대만 가져오기
```

```python
# 공통컬럼과 승차컬럼 연결하기
df_in = pd.concat([df_in1,df_in2], axis=1) # axis=1 -> 옆으로(열기준으로) 합치기
                                            # axis=0 -> 위아래로(행기준으로) 합치기
df_in
```

## 하차 테이블 만들기

```python
# 공통 칼럼
df_out1 = df.iloc[:,:3]

# 하차 컬럼만 가져오기
df_out2 = df.iloc[:,4::2]
df_out2.columns = df_out2.columns.str.split(' ').str[0]

# 공통컬럼과 하차컬럼 연결하기
df_out = pd.concat([df_out1,df_out2], axis=1)
df_out
```
