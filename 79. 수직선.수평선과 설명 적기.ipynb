```python
import pandas as pd
import matplotlib.pyplot as plt
```

# 수평선, 수직선 그리기

## 수평선 그리기
* `plt.axhline(y좌표, x축시작위치, x축끝위치)`<br>
수평선의 길이가 1이라고 했을 때 x축시작위치, x축끝위치를 지정한다.<br>
따로 지정하지 않으면 x축 전범위에 걸쳐 그려진디ㅏ.
* `plt.hlines(y좌표, x축시작좌표, x축끝좌표)`<br>

```python
plt.plot([1,2,3,4], 'ko')

plt.show()
```

```python
plt.plot([1,2,3,4], 'ko')
# plt.axhline(y좌표, x축시작위치, x축끝위치)
plt.axhline(2, 0.1, 0.9, color='r', alpha=0.5, ls=':')
# x축 전체의 비율 = 1

plt.show()
```

```python
plt.plot([1,2,3,4], 'ko')
# plt.hlines(y, x축시작좌표, x축끝좌표)
plt.hlines(3, 1, 2.5, color='g', alpha=0.5, ls='--')
# x축은 좌표값으로 나타냄

plt.show()
```

```python
plt.plot([1,2,3,4], 'ko')

plt.axhline(2, 0.1, 0.9, color='r', alpha=0.5, ls=':')

plt.hlines(3, 1, 2.5, color='g', alpha=0.5, ls='--')

plt.show()
```

## 수직선 그리기
* `plt.axvline(x좌표, y축시작위치, y축끝위치)`<br>
수직선의 길이가 1이라고 했을 때 y축시작위치, y축끝위치를 지정한다.<br>
따로 지정하지 않으면 y축 전범위에 걸쳐 그려진디ㅏ.
* `plt.vlines(x좌표, y축시작좌표, y축끝좌표)`<br>

```python
plt.plot([1,2,3,4], 'ko')

# axvline(x좌표, y축시작위치, y축끝위치)
plt.axvline(1, 0.2, 0.8, color='r', alpha=0.5, ls=':')
# y축은 비율로 계산

plt.show()
```

```python
plt.plot([1,2,3,4], 'ko')

# plt.vlines(x좌표, y축시작좌표, y축끝좌표)
plt.vlines(2, 1.5, 3.5, color='g', alpha=0.5, ls='--')
# y축은 좌표로 계산

plt.show()
```

```python
plt.plot([1,2,3,4], 'ko')

plt.axvline(1, 0.2, 0.8, color='r', alpha=0.5, ls=':')

plt.vlines(2, 1.5, 3.5, color='g', alpha=0.5, ls='--')

plt.show()
```

[학습목표]<br>
그래프의 특정 위치에 설명을 추가할 수 있다.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 그래프에 한글폰트 설정
plt.rcParams['font.family']='Malgun Gothic'

# 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False
```

# 텍스트 추가하기
* plt.text(x좌표, y좌표, 텍스트)
* rotation=회전각도
* ha : horizontal alignment
* va : vertical alignment
* 텍스트 상자<br>
`bbox = {'boxstyle':상자스타일, 'fc':facecolor,'ec':edgecolor,...}`<br>
boxstyle : 'round'/'square'

```python
plt.plot([1,2,3,4], 'ko')

# plt.text(2.1, 3.1, 'test', rotation=320)
# plt.text(2.1, 3, '(x:2, y=3)')
plt.text(2.1, 3, '(x:2, y=3)', ha='left', va='bottom', fontsize=12, rotation=45
        , bbox = {'boxstyle':'round', 'fc':'skyblue','ec':'b', 'alpha':0.3})

plt.axhline(3, ls=':', alpha=0.5, lw=0.5)
plt.axvline(2, ls=':', alpha=0.5, lw=0.5)

plt.plot(2,3,'ro')

plt.show()
```

# 화살표와 텍스트 추가하기
* `plt.annotate('텍스트',xy=(화살표x,화살표y), xytext=(텍스트x,텍스트y), arrowprops=화살표속성(딕셔너리))`<br>
<br>
* 화살표 속성<br>
`width`        The width of the arrow in points<br>
`headwidth`    The width of the base of the arrow head in points<br>
`headlength`   The length of the arrow head in points<br>
`shrink`       Fraction of total length to shrink from both ends<br>

```python
plt.plot([1,2,3,4], 'ko')

plt.axhline(2, color='orange', ls='--', alpha=0.5, lw=0.5)
plt.axvline(1, ls='--', alpha=0.5, lw=0.5, color='orange')

plt.plot(1,2,'ro')


plt.annotate('(x:1, y:2)',xy=(1,2), xytext=(1.5, 2.5)
             , arrowprops={'width':2, 'headwidth':10, 'headlength':10
                           , 'shrink':0.1, 'fc':'g'}
            , fontsize=12, color='r')

plt.show()
```
