# 위치, 크기 지정하여 그래프 그리기

## figure, axes
* figure : 그림이 그려지는 캔버스
* axes : 하나의 그래프
</img>

## 위치, 크기 지정하여 그래프 그리기
1) figure 객체를 생성한다.<br>
<b>`fig = plt.figure(figsize=(가로길이,세로길이))`</b><br><br>
2) figure객체의 add_axes 메소드로 위치와 크기를 지정하여 axes 객체를 생성한다.<br>
<b>`ax1 = fig.add_axes([left, bottom, width, height])`</b><br>
<span style="color:red">left, bottom</span> : 상대적인 시작 위치 (figsize의 크기를 1이라고 했을 때 상대적 위치) <br>
<span style="color:blue">width, height</span> : 상대적인 크기(figsize의 크기를 1이라고 했을 때 상대적 크기) 
<br>
</img>

3) axes에  그래프를 그린다.<br>
<b>`ax1.plot(x,y)`</b>

4) axes에  제목 추가.<br>
<b>`ax1.set_title(제목)`</b>




### 위치와 크기를 자유롭게 지정하여 axes 객체 만들기
* add_axes를 사용하면, 서브플롯의 크기와 위치를 자유롭게 지정할 수 있다.
* 그래프를 겹쳐그리거나, 크기가 각각 다른 그래프를 그릴 수 있다.

### 앤스콤 4분할 그래프 그리기

```python
import seaborn as sns
anscombe = sns.load_dataset('anscombe')
df1 = anscombe[anscombe['dataset']=='I']
df2 = anscombe[anscombe['dataset']=='II']
df3 = anscombe[anscombe['dataset']=='III']
df4 = anscombe[anscombe['dataset']=='IV']
```

```python
df2
```

```python
# 객체 지향으로 그래프 만들기

# 1)figure 객체를 생성한다.
fig = plt.figure(figsize=(8,6))

# 2) figure객체의 add_axes 메소드로 위치와 크기를 지정하여 axes 객체를 생성한다.
# fig.add_axes([left, bottom, width, height])
ax1 = fig.add_axes([0, 0.5, 0.4, 0.4])
ax2 = fig.add_axes([0.5, 0.5, 0.4, 0.4])
ax3 = fig.add_axes([0, 0, 0.4, 0.4])
ax4 = fig.add_axes([0.5, 0, 0.4, 0.4])
#ax5 = fig.add_axes([0.3, 0.3, 0.2, 0.2])

# 3) axes에 그래프를 그린다.
ax1.plot(df1['x'], df1['y'], 'o')
ax2.plot(df2['x'], df2['y'], 'o')
ax3.plot(df3['x'], df3['y'], 'o')
ax4.plot(df4['x'], df4['y'], 'o')

# 4) axes에 제목 추가.
ax1.set_title('ax1')
ax2.set_title('ax2')
ax3.set_title('ax3')
ax4.set_title('ax4')
```

# 데이터 불러오기

```python
import seaborn as sns
anscombe = sns.load_dataset('anscombe')
df1 = anscombe[anscombe['dataset']=='I']
df2 = anscombe[anscombe['dataset']=='II']
df3 = anscombe[anscombe['dataset']=='III']
df4 = anscombe[anscombe['dataset']=='IV']
```

```python
df4
```

# axes를 행, 열로 쪼개어 서브플롯 그리기
* plt.subplots() 함수를 호출하면 figure, axes 객체를 생성하여 튜플 형태로 반환한다.<br>
<b>`fig, ax = plt.subplots()`</b><br><br>
1) axes 객체를 행,열로 쪼개어 생성하기<br>
<b>`fig, ax = plt.subplots(nrows=행개수, ncols=열개수,figsize=(가로사이즈,세로사이즈))`</b><br><br>
2) axes[행번호][열번호] 형태로 접근하여 그래프 그리기
</img>
3) 서브플롯간 축을 공유할 수 있다.<br>
<b>`sharex=True, sharey=True`</b><br><br>

```python
# 1) axes 객체를 행,열로 쪼개어 생성하기
# 3) 서브플롯간 축을 공유할 수 있다.
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(8,6)
                      , sharex=True, sharey=True)

# 2) axes[행번호][열번호] 형태로 접근하여 그래프 그리기
ax[0][0].plot(df1['x'], df1['y'], 'o')
ax[0][1].plot(df2['x'], df2['y'], '^')
ax[1][0].plot(df3['x'], df3['y'], '*')
ax[1][1].plot(df4['x'], df4['y'], 'd')

# 4) 각 그래프에 제목 추가
ax[0][0].set_title('ax1')
ax[0][1].set_title('ax2')
ax[1][0].set_title('ax3')
ax[1][1].set_title('ax4')

# 4) 각 그래프에 그리드 추가
ax[0][0].grid(ls=':')
ax[0][1].grid(ls=':', color='pink')
ax[1][0].grid(ls=':', color='skyblue')
ax[1][1].grid(ls=':', color='green', alpha=0.5)

# 6) 그래프 전체 제목
fig.suptitle('Anscombe', size=20)

# 7) 그래프 간격, 크기 최적화
fig.tight_layout()
```
