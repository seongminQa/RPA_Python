# 스캐터플롯
* 점의 크기, 색깔로 정보를 표현할 수 있다. (버블차트)
* 크기는 s, 색깔은 c로 지정한다.

```python
# 데이터
x = [1,5,6,9,10]
y = [1,5,3,9,7]
```

## 기본

```python
plt.scatter(x,y)
# 점의 크기와 색을 더 자세히 바꿀 수 있다.
```

```python
plt.plot(x,y,'o')
```

## 점의 크기
<b>`s=크기공통지정`</b><br>
<b>`s=크기목록`</b>
* plt.scatter()의 파라미터 s로 점 크기를 지정한다.
* 점 크기는 공통된 크기로 지정할 수도 있고, 점마다 다른 크기로 지정할 수도 있다.<br>


```python
plt.scatter(x,y,s=[100,200,50,80,30])
```

## 점의 색상
<b>`c=색상공통지정`</b><br>
<b>`c=색상목록`</b>
* plt.scatter()의 파라미터 c로 점 색상을 지정한다.
* 점 색상은 공통된 색상으로 지정할 수도 있고, 점마다 다른 색상으로 지정할 수도 있다.<br>

### 단일색 지정

```python
plt.scatter(x,y,s=[100,200,50,80,30], c='r')
```

### 값에 따른 색상 지정
* ['r','b','g','k','y']

```python
plt.scatter(x,y,s=[100,200,50,80,30], c=['r','b','g','k','y'])
```

### 컬러맵으로 색상 지정
* 컬러맵 지정 : cmap=컬러맵
* 컬러바 표시 : plt.colorbar()  
* 컬러맵 : https://matplotlib.org/3.3.1/tutorials/colors/colormaps.html

```python
plt.scatter(x,y,c=y, cmap='cool')
plt.colorbar() 
# y값 수치에 따른 컬러바의 변화
```

```python
# 컬러맵 종류
plt.colormaps()
# 컬러맵 리스트들 확인가능
```

# 산점도 예제

## 데이터 준비
* 어느 레스토랑의 팁 데이터
* 지불금액, 팁, 성별, 흡연여부, 요일, 시간대, 테이블인원

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

## 지불금액과 팁의 관계
* 지불금액이 클수록 팁도 많이 줄까?

```python
plt.scatter(tips['total_bill'], tips['tip'])
```

## 점의 크기로 테이블 인원 표시

```python
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*50, alpha=0.5)
# 알파는 투명도
```

## 점의 색으로 성별 표시

```python
# 성별에 따른 color 컬럼 추가
def set_color(x):
    if x=='Male':
        return 'blue'
    elif x=='Female':
        return 'red'
    
tips['color'] = tips['sex'].apply(set_color)
tips
```

```python
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*50, alpha=0.5,
           c=tips['color'])
```

# 히트맵 예제
* 타이나틱호의 연령대_객실등급별 승선자수

## 데이터 준비

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic
```

## 데이터 전처리 

### 결측치 처리

```python
# 결측치 확인
titanic.isnull().sum()
```

```python
# 컬럼 삭제
titanic = titanic.drop(columns=['deck'])
```

```python
# 결측치 삭제
titanic = titanic.dropna()
```

```python
titanic.isnull().sum()
```

```python
titanic.shape
```

### 연령대 컬럼 생성

```python
titanic['agerange'] = (titanic['age']/10).astype('int')*10
```

```python
titanic
```

### 피벗테이블 : 연령대-객실등급 별 승선자 수

```python
titanic_pivot = titanic.pivot_table(index='class', 
                    columns='agerange', values='survived', aggfunc='count')
```

```python
titanic_pivot
```

## 히트맵

### matplotlib

```python
plt.pcolor(titanic_pivot)
plt.colorbar()
# x,y축의 tick이 제대로 되어있지 않다.
```

```python
plt.pcolor(titanic_pivot)
plt.colorbar()

plt.xticks(range(0, len(titanic_pivot.columns),1) )
# len(titanic_pivot.columns) = 9

plt.yticks(range(0, len(titanic_pivot.index),1))

plt.show()
```

```python
plt.pcolor(titanic_pivot)
plt.colorbar()

plt.xticks(np.arange(0.5, len(titanic_pivot.columns), 1), 
           labels=titanic_pivot.columns )

plt.yticks(np.arange(0.5, len(titanic_pivot.index), 1), 
           labels=titanic_pivot.index )

plt.show()
```

### seaborn
* heatmap(data=2차원데이터)<br>
cmap=컬러맵 : 컬러맵 지정<br>
annot=True : 수치 표시<br>
fmt='d' : 정수로 표시<br>

```python
# 히트맵 -> seaborn 으로 더 쉽게 그릴 수 있다
sns.heatmap(titanic_pivot, cmap='Blues', annot=True, fmt='d')
```
