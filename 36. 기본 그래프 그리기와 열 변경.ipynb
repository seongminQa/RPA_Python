```python
import pandas as pd
df = pd.read_csv('data/scores.csv')
df.head()
```

# 열 추가/수정하기
- 데이터프레임[컬럼] = 추가할 데이터
- 데이터프레임[컬럼] = 수정할 데이터
-- 컬럼이 존재하면 추가, 존재하지 않으면 수정된다

## 열 추가하기

```python
# 학생 번호 추가하기(1부터 시작하여 1씩 증가)
df['no'] = range(1,len(df)+1)
df
```

```python
# sum 추가
df['sum'] = df['kor']+df['eng']+df['math']
df
```

## 열 수정하기

```python
# 학생 번호 수정하기 (100부터 시작해서 1씩 증가)
df['no'] = df['no'] + 99
df
```

# 열 삭제하기
- 데이터프레임.drop(columns=삭제할컬럼리스트,inplace=True)
- 존재하지 않는 열은 삭제할 수 없다

```python
# no, sum 컬럼 삭제하기
df.drop(columns=['no','sum']) # -> no, sum이 삭제된 것을 미리보기로 알려준다.
df.drop(columns=['no','sum'],inplace=True) # inplace=True를 적어줘야 실제로 적용된다) # inplace=True를 적어줘야 실제로 적용된다
df

# 또는
# df = df.drop(columns=['no','sum']) 이렇게 해주어도 삭제된다.
```

# 컬럼명 바꾸기

## 컬럼명 한번에 바꾸기
- 데이터프레임.columns = 컬럼명리스트
-- 컬럼명리스트의 항목 수는 컬럼 수와 동일해야 한다

```python
# '이름','국어','영어','수학'
df.columns = ['이름','국어','영어','수학']
df.head(5)
```

```python
# 전체컬럼수와 컬럼명 리스트의 항목 수가 다르면 error
```

```python
# 이름 --> 성명
df.rename(columns={'이름':'성명'}) # 미리보기 된것 df
df= df.rename(columns={'이름':'성명'})
df
```
